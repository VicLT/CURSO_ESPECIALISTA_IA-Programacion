#Q7. Sum of weights
1 - Select all the cars from Asia.

```python
asia_df = df[df['origin'].str.lower() == 'asia']
asia_df.head()
```

2 - Select only columns vehicle_weight and model_year.

```python
asia_features = asia_df[['vehicle_weight', 'model_year']]
asia_features.head()
```

3 - Select the first 7 values.

```python
asia_top7 = asia_features.head(7)
asia_top7.head()
```

4 - Get the underlying NumPy array. Let's call it X.

```python
#X: Matriz de características (7 filas × 2 columnas)
X = asia_top7.to_numpy()
```

5 - Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.

```python
#XTX: Producto de X transpuesta por X (2×2)
XTX = X.T @ X
```

6 - Invert XTX.

```python
import numpy as np

#XTX_inv: Inversa de XTX
XTX_inv = np.linalg.inv(XTX)
```

7 - Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].

```python
#y: Vector de 7 valores
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

8 - Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.

```python
#w: Resultado del cálculo de regresión lineal por fórmula cerrada (método de mínimos cuadrados)
w = XTX_inv @ X.T @ y
```

9 - What's the sum of all the elements of the result?

```python
#Suma de los coeficientes
result_sum = w.sum()

print(f"Sum of all elements in w: {result_sum}")
```

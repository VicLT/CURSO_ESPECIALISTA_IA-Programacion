### Manos en la masa 1!

Usando el módulo `random`, desarrolla un bombo  que pregunta al usuario el tamaño del bombo y el número de números que quiere extraer y los presenta en pantalla


````
import random

cantidad_numeros_in = float(input("Introduzca tamaño del bombo: "))
cantidad_numeros_in = int(cantidad_numeros_in)

cantidad_numeros_out = float(input("Introduzca la cantidad de números que desea extraer: "))
cantidad_numeros_out = int(cantidad_numeros_out)

bombo = list(range(1, cantidad_numeros_in + 1))
print(f"Números del bombo: {bombo}.")
print(f"Desordenando números...")
random.shuffle(bombo)
print(bombo)

for i in range(0, cantidad_numeros_out):
  print(f"El número elegido en la posición {i+1} es {bombo[i]}.")
````

````
import zlib

texto = "l6OCEQQZgOFBGs4R1oZLtuYNsGOdtgemk4juxbss8M5jMMlwBp9rgRVN7w15Hp89s5obROPCpRJBElWgOcPl9F2koApY9DNtLui87HMyk0X5CmywYwQPGzMOmMJuupixGZhyv0Fth3slL7VBdqLb7a96yZDsHLVmlh96EWN0N1wKFv"
texto = 100 * texto

texto_bytes = texto.encode("utf-8")
print(f"Tamaño para la cadena no comprimida: {len(texto_bytes)} B.")

texto_comprimido = zlib.compress(texto_bytes)
print(f"Tamaño para la cadena comprimida: {len(texto_comprimido)} B.")

print(f"Eficiencia de compresión: {len(texto_comprimido)/len(texto_bytes)}")
````

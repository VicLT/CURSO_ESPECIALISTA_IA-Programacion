###Manos en la masa 3!

Divide el siguiente verso de Miguel Hernández en una lista de palabras:


`````
frase = "Alto soy de mirar a las palmeras, rudo de convivir con las montañas..."
`````

Entonces, cuenta el número de palabras de tres letras en esta lista.

`````
frase = "Alto soy de mirar a las palmeras, rudo de convivir con las montañas..."
palabras = frase.split()
palabras_de_tres_letras = []

for palabra in palabras:
  if len(palabra) == 3:
    palabras_de_tres_letras.append(palabra)

print(palabras_de_tres_letras)
print(f"En la frase hay {len(palabras_de_tres_letras)} palabras de tres letras.")
`````

###Manos en la Masa 5!

Dado el poema Aceituneros de Miguel Hernández:



`````
aceituneros='''Andaluces de Jaén,
aceituneros altivos,
decidme en el alma: ¿quién,
quién levantó los olivos?
No los levantó la nada,
ni el dinero, ni el señor,
sino la tierra callada,
el trabajo y el sudor.
Unidos al agua pura
y a los planetas unidos,
los tres dieron la hermosura
de los troncos retorcidos.
Levántate, olivo cano,
dijeron al pie del viento.
Y el olivo alzó una mano
poderosa de cimiento.
Andaluces de Jaén,
aceituneros altivos,
decidme en el alma: ¿quién
amamantó los olivos?
Vuestra sangre, vuestra vida,
no la del explotador
que se enriqueció en la herida
generosa del sudor.
No la del terrateniente
que os sepultó en la pobreza,
que os pisoteó la frente,
que os redujo la cabeza.
Árboles que vuestro afán
consagró al centro del día
eran principio de un pan
que sólo el otro comía.
¡Cuántos siglos de aceituna,
los pies y las manos presos,
sol a sol y luna a luna,
pesan sobre vuestros huesos!
Andaluces de Jaén,
aceituneros altivos,
pregunta mi alma: ¿de quién,
de quién son estos olivos?
Jaén, levántate brava
sobre tus piedras lunares,
no vayas a ser esclava
con todos tus olivares.
Dentro de la claridad
del aceite y sus aromas,
indican tu libertad
la libertad de tus lomas.'''
`````

Diseña un programa para encontrar si la palabra  `aceituneros`, una palabra clave específica puede ser encontrada en los datos.

`````
palabra_clave = "aceituneros"
palabras = aceituneros.split()
cantidad_de_palabras_clave = 0

for palabra in palabras:
    if palabra == palabra_clave:
        cantidad_de_palabras_clave += 1

if cantidad_de_palabras_clave > 0:
    print(f"La palabra '{palabra_clave}' ha sido encontrada {cantidad_de_palabras_clave} veces.")
else:
    print(f"La palabra '{palabra_clave}' no se ha podido encontrar.")
`````

###Manos en la masa 6!

Dado un frase diseñar un programa para eliminar todas las vocales de esa cadena de texto

`````
texto = "Las investigaciones han demostrado que a menudo todavía es posible entender el texto incluso si se eliminan todas las vocales."
vocales = ["a", "á", "e", "é", "i", "í", "o", "ó", "u", "ú"]

for vocal in vocales:
    texto = texto.replace(vocal, "")

print(texto)
`````

###Manos en la masa 8!

Cuenta el número de veces que ocurre alguna forma de *bla*, independientemente del caso.

`````
palabras = ["BlA", "plop", "blA", "Bla", "PLoP", "bLa", "pLop", "BLA"]
palabra_clave = "bla"
contador_bla = 0

for palabra in palabras:
    if palabra.lower() == palabra_clave:
        contador_bla += 1

print (f"Cantidad de 'bla': {contador_bla}")
`````

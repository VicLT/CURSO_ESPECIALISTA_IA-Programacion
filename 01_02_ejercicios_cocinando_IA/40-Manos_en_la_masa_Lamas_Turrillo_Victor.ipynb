##La función enumerate()

###Manos en la masa 1!

````
obras_mh = ["Nanas de la cebolla", "Viento del Pueblo", "Elegía"]
````

sustituye `# 1`, `# 2`, `# 3` por `["primer","segundo","tercer"]` para que la salida del programa sea
```
El primer libro de Miguel Hernández es Nanas de la cebolla
El segundo libro de Miguel Hernández es Viento del Pueblo
El tercer de Miguel Hernández es Elegía
```
usando solo una linea más de programación y cambiando lo que necesites en el resto

````
orden = ["primer", "segundo", "tercer"]
for posicion, libro in enumerate(obras_mh):
    print(f"El {orden[posicion]} libro de Miguel Hernández es '{libro}'.")
````

##La función zip()

### Manos en la masa 2!

En el ejemplo anterior hemos usamos dos listas, pero también podemos reescribir todo este código para usar un diccionario, muy adecuado para tal tarea. Usa un bucle para iterar sobre las claves del diccionario y hazlo también usando el método `.items()`de los diccionarios. ¡Inténtalo!


````
dic_obras_mh = {"Nanas de la cebolla":9, "Viento del Pueblo":8, "Elegía":7}

for libro in dic_obras_mh:
    print(f"El libro '{libro}' tiene una valoración de {dic_obras_mh[libro]}.")
````

````
for libro, valoracion in dic_obras_mh.items():
    print(f"El libro '{libro}' tiene una valoración de {valoracion}.")
````

## El bucle `while`

### Manos en la masa 5!

Vamos a sumar los valores entre dos valores introducidos por el usuario en pantalla:

````
cota_inferior = int(input("Introduce el valor inferior: "))
cota_superior = int(input("Introduce el valor superior: "))
suma = 0

if cota_inferior > cota_superior:
    print("Vuelve a introducir los datos correctamente.")
else:
    for k in range(cota_inferior, cota_superior + 1):
        suma += k

print(suma)
````

### Manos en la masa 6!

Imprimir qué números son pares e impares en un rango dado utilizando el operador % que da el resto de la división `10%2==0` (par) `11%2==1` (impar)

````
numero1 = int(input("Escriba un número entero: "))
numero2 = int(input(f"Escriba un número entero igual o mayor que {numero1}: "))

if numero2 < numero1:
    print(f"¡Le he pedido un número entero igual o mayor que {numero1}!")
else:
    for num in range(numero1, numero2 + 1):
        if num % 2 == 0:
            print(f"El número {num} es par.")
        else:
            print(f"El número {num} es impar.")
````

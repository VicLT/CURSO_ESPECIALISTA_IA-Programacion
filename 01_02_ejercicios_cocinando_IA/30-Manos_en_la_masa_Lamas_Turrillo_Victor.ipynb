###Manos en la masa 1!

Vamos a practicar nuestras nuevas habilidades de condición un poco. Escribe un pequeño programa que defina la variable `peso`. Si `peso` es > 23 kilos, imprimimos "Hay un cargo adicional de 25 dólares por un equipaje tan pesado". Si no lo es, imprimimos: "Gracias por su viaje". Cambia el valor del peso para ver los dos ejemplos. (Utiliza los operadores `<` o `>`)


```
import random

peso = random.randint(1, 50)
print(f"Peso: {peso}")

if peso > 23:
  print("Hay un cargo adicional de 25$ por un equipaje tan pesado")
else:
  print("Gracias por su viaje.")
```

###Manos en la masa 2!

¿Puedes escribir un programa que imprima "Esta es un lista vacia" si la lista dada no contiene ningún valor?

```
lista_vacia = []

if not lista_vacia:
  print("Esta es una lista vacía.")
```

¿Puedes hacer esto mismo pero con la función `len()`?


```
lista = [1, 2, 3]

if len(lista) == 0:
    print("Esta es una lista vacía.")
else:
    print("La lista no está vacía.")
```

###Manos en la masa 3!

Ahora escribe un pequeño programa que itere a lo largo de la lista `colores`y añada (`appends`) todos los colores que contengan la letra *r* a la lista `colores_con_r` y la imprima:


```
colores = ["amarillo", "rojo", "verde", "azul", "púrpura"]
colores_con_r = []
letra = "r"

for palabra in colores:
  if letra in palabra:
    colores_con_r.append(palabra)
    print(palabra)
```

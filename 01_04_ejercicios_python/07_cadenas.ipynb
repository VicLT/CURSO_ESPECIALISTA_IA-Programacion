##1. Pide al usuario una frase y cuenta cuántas vocales tiene.

##2. Pide una palabra y comprueba si es un palíndromo (igual al revés).

```python
palabra = str(input("Introduzca una palabra: "))
caracteres = []
caracteres_inv = []

for i in range(len(palabra)):
    caracteres.append(palabra[i])
    caracteres_inv.append(palabra[-(i + 1)])

if len(palabra) > 0:
  if caracteres == caracteres_inv:
      print(f"\n'{palabra}' es un palíndromo.")
  else:
      print(f"\n'{palabra}' no es un palíndromo.")
else:
  print("\nEntrada inválida. Por favor, introduzca una palabra.")
```

##3. Dada una frase, elimina los espacios de los extremos y conviértela en formato título (title()).

##4. Divide una cadena en palabras y muestra cuántas hay.

```python
frase = input("Introduzca una frase: ")

palabras = frase.split()
cantidad_palabras = len(palabras)

print(f"\nLa frase contiene {cantidad_palabras} palabra(s).")
```

###5. Escribe una función que reciba una cadena y devuelva cuántas veces aparece cada vocal.

```python
def contar_vocales(cadena):
    cadena = cadena.lower()

    contador = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}

    for letra in cadena:
        if letra in contador:
            contador[letra] += 1

    return contador

texto = input("Introduzca una frase: ")
contador_vocales = contar_vocales(texto)

print(f"\nCantidad vocales:")
for vocal, cantidad in contador_vocales.items():
    print(f"{vocal}: {cantidad}")
```

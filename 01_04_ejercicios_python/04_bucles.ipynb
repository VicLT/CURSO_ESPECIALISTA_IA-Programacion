##1. Imprime los números del 1 al 10 usando for.

````python
for numero in range(1, 11):
    print(numero)
````

##2. Imprime los pares del 2 al 20.

````python
for numero in range(2, 21):
    if numero % 2 == 0:
        print(numero)
````

##3. Recorre una lista de frutas y muestra cada una.

````python
frutas = ['manzana', 'pera', 'melocotón']
for fruta in frutas:
    print(fruta)
````

##5. Recorre la palabra "programar" y cuenta cuántas veces aparece la letra "r".

````python
palabra = "programar"
cantidad_r = 0

for letra in palabra:
    if letra == 'r':
        cantidad_r += 1

print(f"La 'r' en la palabra '{palabra}' aparece {cantidad_r} veces.")
````

##6. Haz un programa que pida al usuario un número n y dibuje una pirámide con ceros así:
```python
n = 4
0

```

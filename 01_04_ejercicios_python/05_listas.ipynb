##1.	Crea una lista con los nombres de 5 amigos y recórrela con un for.

```python
amigos = ['Pepe', 'Luisa', 'Mateo', 'Rubén', 'Beatriz']

for nombre in amigos:
    print(nombre)
```

##2.	Pide 5 números al usuario, guárdalos en una lista y muestra su media.

```python
numeros = []
media = 0

for numero in range(1, 6):
    numeros.append(int(input("Introduzca un número: ")))

media = sum(numeros) / len(numeros)
print(f"La media de la lista es {media}.")
```

##3.	Crea una lista de números del 1 al 10. Elimina los pares.

```python
numeros_impares = []

for numero in range(1, 11):
    numeros_impares.append(numero)

for numero in numeros_impares:
    if numero % 2 == 0:
        numeros_impares.remove(numero)

print(numeros_impares)
```

##4.	Dada una lista de palabras, muestra cuántas tienen más de 5 letras.

```python
palabras = ['hola', 'adiós', 'eureka', 'sms', 'instituto']
palabras_con_mas_de_5_letras = 0

for palabra in palabras:
    if len(palabra) > 5:
        palabras_con_mas_de_5_letras += 1

print(f"La cantidad de palabras con más de 5 letras es {palabras_con_mas_de_5_letras}.")
```

##5.	Crea una lista de temperaturas, y calcula la máxima, mínima y promedio.

```python
temperaturas = [20, 25, 30, 35, 40, 45]
maxima = 0
minima = 0
promedio = 0

maxima = max(temperaturas)
minima = min(temperaturas)
promedio = sum(temperaturas) / len(temperaturas)

print(f"La temperatura máxima es {maxima}, la mínima es {minima} y la promedio es {promedio}.")
```

##6. Haz un programa que lea números introducidos por el usuario hasta que escriba "fin" y luego muestre:
*   Cuántos números se han introducido.
*   Su suma y su media.
*   El número más alto y el más bajo.



```python
numeros = []
exit = False

while not exit:
  valor = input("Introduzca un valor numérico o escriba 'fin': ")
  if valor.lower() == "fin":
    exit = True
  else:
    try:
      numeros.append(int(valor))
    except ValueError:
      print("Entrada inválida. Por favor, introduzca un número o 'fin'.")

if numeros:
  cantidad_numeros = len(numeros)
  suma = sum(numeros)
  media = suma / cantidad_numeros
  numero_mas_alto = max(numeros)
  numero_mas_bajo = min(numeros)

  print(f"\nSe han introducido {cantidad_numeros} valores.")
  print(f"Suma = {suma}")
  print(f"Media = {media}")
  print(f"Valor más alto = {numero_mas_alto}")
  print(f"Valor más bajo = {numero_mas_bajo}")
else:
  print("\nNo se introdujeron números.")
```

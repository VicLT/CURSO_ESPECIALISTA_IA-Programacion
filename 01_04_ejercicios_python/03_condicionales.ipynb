#Módulo 3: Condicionales en Python

1. Pide una edad al usuario y muestra si puede votar (mayor o igual a 18).

```python
try:
  edad = int(input("Introduzca su edad: "))
  if edad >= 18:
    print(f"\nEres mayor de edad. Puedes votar.")
  elif edad < 18:
    print(f"\nEres menor de edad. No puedes votar.")
except ValueError:
  print(f"\nEntrada inválida. Por favor, introduzca un valor numérico.")
```

2. Pide una nota y clasifícala como:

*   Sobresaliente (≥ 9)
*   Notable (≥ 7)
*   Aprobado (≥ 5)
*   Suspenso (< 5)

```python
try:
  nota = int(input("Introduzca nota: "))
  if nota >= 9:
    print(f"\nSobresaliente.")
  elif nota >= 7 and nota < 9:
    print(f"\nNotable.")
  elif nota >= 5 and nota < 7:
    print(f"\nAprobado.")
  else:
    print(f"\nSuspenso.")
except ValueError:
  print(f"\nEntrada inválida. Por favor, introduzca un valor numérico.")
```

3. Pide dos números y di cuál es mayor, menor o si son iguales.

```python
try:
  num1 = int(input("Introduzca primer número: "))
  num2 = int(input("Introduzca segundo número: "))

  if num1 == num2:
    print(f"{num1} es igual que {num2}.")
  elif num1 > num2:
    print(f"{num1} es mayor que {num2}.")
  else:
    print(f"{num1} es menor que {num2}.")
except ValueError:
  print(f"\nEntrada inválida. Por favor, introduzca un valor numérico.")
```

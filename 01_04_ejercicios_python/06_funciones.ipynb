##4.	Haz una función que reciba una lista y devuelva la media.

```python
def calcular_media(numeros):
    if len(numeros) > 0:
        media = sum(numeros) / len(numeros)
        return(print(f"\nLa media de {numeros} es {media}."))
    else:
        return(print(f"\nLa lista de números está vacía."))

numeros = []
exit = False

while not exit:
  entrada = input("Introduzca un valor numérico ['exit' para salir]: ")
  if entrada.lower() == "exit":
    exit = True
  else:
    try:
      numeros.append(int(entrada))
    except ValueError:
      print("Entrada inválida. Por favor, introduzca un número o 'exit'.")

calcular_media(numeros)
```

##5.	Crea una función que reciba una palabra y devuelva cuántas vocales tiene.

```python
def contar_vocales(palabra):
    vocales = 0

    for letra in palabra:
        if letra == 'a' or letra == 'e' or letra == 'i' or letra == 'u':
            vocales += 1

    if len(palabra) > 1:
      return(print(f"\nLa palabra '{palabra}' contiene {vocales} vocales."))
    else:
      return(print(f"\nNo se ha introducido ninguna palabra."))

entrada = input("Introduzca una palabra: ")

contar_vocales(entrada)
```
